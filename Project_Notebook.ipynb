## Eliminar Llaves primarias y columnas innecesarias

```python
df = df.drop(columns={"NIT", "RAZÓN SOCIAL", "DEPARTAMENTO DOMICILIO","CIUDAD DOMICILIO", "MACROSECTOR", "Año de Corte"})
df.head(5)
```

## Encontrar y tratar datos nulos

```python
#Encontrar el total de nulos en todo el dataframe
df.isnull().sum()
```

## Eliminar filas de baja frecuencia

```python
category_counts = df['REGIÓN'].value_counts()
max_count = category_counts.max()
threshold = max_count * 0.1
valid_categories = category_counts[category_counts > threshold].index
df_filtered = df[df['REGIÓN'].isin(valid_categories)]
df_filtered['CIIU'] = df_filtered['REGIÓN'].astype(str)
plt.hist(df_filtered['REGIÓN'])
plt.title('Frecuencia REGIÓN')
plt.xlabel('REGIÓN')
plt.ylabel('Frecuencia')
plt.show()
```

```python
category_counts = df['SUPERVISOR'].value_counts()
max_count = category_counts.max()
threshold = max_count * 0.05
valid_categories = category_counts[category_counts > threshold].index
df_filtered = df[df['SUPERVISOR'].isin(valid_categories)]

plt.hist(df_filtered['SUPERVISOR'])
plt.title('Frecuencia SUPERVISOR')
plt.xlabel('SUPERVISOR')
plt.ylabel('Frecuencia')
plt.show()
```

## Tratamiendo de filas numericas (Transformar objects a int o float) y dummies

```python
df_encoded = pd.get_dummies(df_filtered, columns=['SUPERVISOR', 'REGIÓN'])
df_encoded.head(5)
```

```python
df_encoded['INGRESOS OPERACIONALES'] = df_encoded['INGRESOS OPERACIONALES'].replace('[\$,]', '', regex=True).astype(float)
df_encoded['GANANCIA (PÉRDIDA)'] = df_encoded['GANANCIA (PÉRDIDA)'].replace('[\$,]', '', regex=True).astype(float)
df_encoded['TOTAL ACTIVOS'] = df_encoded['TOTAL ACTIVOS'].replace('[\$,]', '', regex=True).astype(float)
df_encoded['TOTAL PASIVOS'] = df_encoded['TOTAL PASIVOS'].replace('[\$,]', '', regex=True).astype(float)
df_encoded['TOTAL PATRIMONIO'] = df_encoded['TOTAL PATRIMONIO'].replace('[\$,]', '', regex=True).astype(float)

df_encoded['CIIU'] = df_encoded['CIIU'].astype(int)
df_encoded.head(5)
```

## Eliminar valores que son 0 que bajo contexto son nulos

```python
## Elimar valores que son 0

columns_clean = ["INGRESOS OPERACIONALES", "GANANCIA (PÉRDIDA)", "TOTAL ACTIVOS", "TOTAL PASIVOS", "TOTAL PATRIMONIO"]

df_encoded[columns_clean] = df_encoded[columns_clean].replace(0, pd.NA)
df_encoded = df_encoded.dropna()
df_encoded.head(5)
```

## Calcular indices de riesgo financiero

```python
#Ratio de Liquidez - "Un ratio de liquidez alto indica que la empresa tiene suficientes activos para cubrir sus pasivos"
df_encoded["RATIO_LIQUIDEZ"] = df_encoded["TOTAL ACTIVOS"] / df_encoded["TOTAL PASIVOS"]

#Ratio Endeudamiento - "Un ratio de endeudamiento alto puede ser un signo de riesgo, debido a la cantidad de deudas"
df_encoded["RATIO_ENDEUDAMIENTO"] = df_encoded["TOTAL PASIVOS"] / df_encoded["TOTAL PATRIMONIO"]

#Margen Bruto - 
#"Un margen bruto alto indica que la empresa tiene una buena eficiencia en la conversión de sus ingresos operacionales en ganancias"
df_encoded["MARGEN_BRUTO"] = df_encoded["GANANCIA (PÉRDIDA)"] / df_encoded["INGRESOS OPERACIONALES"]

#ROA (Ratio de Retorno sobre Activos) - "Un ROA alto indica que la empresa está generando ganancias de manera eficiente"
df_encoded["ROA"] = df_encoded["GANANCIA (PÉRDIDA)"] / df_encoded["TOTAL ACTIVOS"]
```

```python
df_encoded.fillna(0, inplace=True)
df_encoded.replace([np.inf, -np.inf], 0, inplace=True)
df_encoded
```

```python
#Visualizar correlaciones antes de crear indice general
import seaborn as sns

correlaciones = df_encoded.corr()
plt.figure(figsize=(10,8)) 
sns.heatmap(correlaciones, cmap="coolwarm") 
plt.show()
```

```python
#Crear Indice General que reemplace al CVI 
peso = [0.3, 0.2, 0.2, 0.3] 
df_encoded['Indice_General'] = df_encoded["RATIO_LIQUIDEZ"] * peso[0] - df_encoded['RATIO_ENDEUDAMIENTO'] * peso[1] + df_encoded['MARGEN_BRUTO'] * peso[2] + df_encoded['ROA'] * peso[3]
df_encoded
```

## Normalizar las columnas numericas

```python
df_encoded = df_encoded.astype(float)
columns = df_encoded.columns.tolist()
for col in columns:
    print(f"Processing column {col}....")
    df_encoded[col] = df_encoded[col].apply(lambda x: (x - df_encoded[col].min()) / (df_encoded[col].max() - df_encoded[col].min()))
                            
df_encoded.head(5)
```

## Modelos de prediccion Indice General

### Random Forest Model

```python
# Split del datafrme 
from sklearn.model_selection import train_test_split

y = df_encoded['Indice_General']
columnas_input = ["INGRESOS OPERACIONALES", "GANANCIA (PÉRDIDA)", "TOTAL ACTIVOS", "TOTAL PASIVOS", "TOTAL PATRIMONIO"]
x = df_encoded[columnas_input]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
#Tamaños de los splits
print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
#Inicializar el random forest con 100 arboles, y una maxima profundidad de 10 niveles, semilla random 42
random_forest_forest = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
random_forest_forest.fit(X_train, y_train)
```

```python
#Predicciones del modelo
y_pred = random_forest_forest.predict(X_test)
y_pred
```

### Metricas del Modelo Random Forest

```python
#Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error: ", mse)
```

#### Conclusiones

MSE hace referencia a la media del error al cuadrado, lo cual hace que en este indicador se vuelva mas sensible a errores muy grandes. Tener una media baja implica que el modelo no comete tantos errores

```python
#Root Mean Squared Error (RMSE)
import numpy as np
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error: ", rmse)
```

#### Conclusiones
Es la raiz cuadrada del MSE, lo cual implica que tiene las misma escala que la variable target. Un bajo RMSE es favorable

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error: ", mae)
```

#### Conclusiones

El MAE se refiere a la media del error absoluto en las predicciones. Un error bajo es favorable

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R-squared: ", r2)
```

#### Conclusiones
Mas de la mitad de los datos pueden ser explicados o representados por el modelo teniendo en cuenta el valor del r^2

### Regresión Logistica

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from matplotlib.colors import ListedColormap
```

```python
#Input del modelo
listax = ["INGRESOS OPERACIONALES","GANANCIA (PÉRDIDA)","TOTAL ACTIVOS","TOTAL PASIVOS","TOTAL PATRIMONIO"]
x = df_encoded[listax]
#Output
y = df_encoded['Indice_General']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    x, y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import Binarizer

# Binarizar variables
binarizer = Binarizer(threshold=0.5)
y_bin = binarizer.transform(ytrain.values.reshape(-1, 1))
ytest_bin = binarizer.transform(ytest.values.reshape(-1, 1))

regresion = LogisticRegression(random_state=0)
regresion.fit(xtrain, y_bin)

```

```python
y_bin
```

```python

y_pred = regresion.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(ytest_bin, y_pred)
mse = mean_squared_error(ytest_bin, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(ytest_bin, y_pred)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'R2 Score: {r2}')
```

## Modelos de cluster para prediccion de CIIU

### K-Medias

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6) # You can change the number of clusters
kmeans.fit(df_encoded[["Indice_General"]])
```

```python

print(kmeans.cluster_centers_)
```

```python

df_encoded['cluster'] = kmeans.predict(df_encoded[['Indice_General']])
```

```python

plt.scatter(df_encoded['Indice_General'], df_encoded['CIIU'], c=df_encoded['cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=50)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(df_encoded[['Indice_General']], kmeans.labels_)
print("Silhouette Score: ", score)
```
