## Eliminar Llaves primarias y columnas innecesarias

```python
df = df.drop(columns={"NIT", "RAZÓN SOCIAL", "DEPARTAMENTO DOMICILIO","CIUDAD DOMICILIO", "MACROSECTOR", "Año de Corte"})
df.head(5)
```

## Encontrar y tratar datos nulos

```python
#Encontrar el total de nulos en todo el dataframe
df.isnull().sum()
```

## Eliminar filas de baja frecuencia

```python
category_counts = df['REGIÓN'].value_counts()
max_count = category_counts.max()
threshold = max_count * 0.1
valid_categories = category_counts[category_counts > threshold].index
df_filtered = df[df['REGIÓN'].isin(valid_categories)]
df_filtered['CIIU'] = df_filtered['REGIÓN'].astype(str)
plt.hist(df_filtered['REGIÓN'])
plt.title('Frecuencia REGIÓN')
plt.xlabel('REGIÓN')
plt.ylabel('Frecuencia')
plt.show()
```

```python
category_counts = df['SUPERVISOR'].value_counts()
max_count = category_counts.max()
threshold = max_count * 0.05
valid_categories = category_counts[category_counts > threshold].index
df_filtered = df[df['SUPERVISOR'].isin(valid_categories)]

plt.hist(df_filtered['SUPERVISOR'])
plt.title('Frecuencia SUPERVISOR')
plt.xlabel('SUPERVISOR')
plt.ylabel('Frecuencia')
plt.show()
```

## Tratamiendo de filas numericas (Transformar objects a int o float) y dummies

```python
df_encoded = pd.get_dummies(df_filtered, columns=['SUPERVISOR', 'REGIÓN'])
df_encoded.head(5)
```

```python
df_encoded['INGRESOS OPERACIONALES'] = df_encoded['INGRESOS OPERACIONALES'].replace('[\$,]', '', regex=True).astype(float)
df_encoded['GANANCIA (PÉRDIDA)'] = df_encoded['GANANCIA (PÉRDIDA)'].replace('[\$,]', '', regex=True).astype(float)
df_encoded['TOTAL ACTIVOS'] = df_encoded['TOTAL ACTIVOS'].replace('[\$,]', '', regex=True).astype(float)
df_encoded['TOTAL PASIVOS'] = df_encoded['TOTAL PASIVOS'].replace('[\$,]', '', regex=True).astype(float)
df_encoded['TOTAL PATRIMONIO'] = df_encoded['TOTAL PATRIMONIO'].replace('[\$,]', '', regex=True).astype(float)

df_encoded['CIIU'] = df_encoded['CIIU'].astype(int)
df_encoded.head(5)
```

## Calcular indices de riesgo financiero

```python
#Ratio de Liquidez - "Un ratio de liquidez alto indica que la empresa tiene suficientes activos para cubrir sus pasivos"
df_encoded["RATIO_LIQUIDEZ"] = df_encoded["TOTAL ACTIVOS"] / df_encoded["TOTAL PASIVOS"]

#Ratio Endeudamiento - "Un ratio de endeudamiento alto puede ser un signo de riesgo, debido a la cantidad de deudas"
df_encoded["RATIO_ENDEUDAMIENTO"] = df_encoded["TOTAL PASIVOS"] / df_encoded["TOTAL PATRIMONIO"]

#Margen Bruto - 
#"Un margen bruto alto indica que la empresa tiene una buena eficiencia en la conversión de sus ingresos operacionales en ganancias"
df_encoded["MARGEN_BRUTO"] = df_encoded["GANANCIA (PÉRDIDA)"] / df_encoded["INGRESOS OPERACIONALES"]

#ROA (Ratio de Retorno sobre Activos) - "Un ROA alto indica que la empresa está generando ganancias de manera eficiente"
df_encoded["ROA"] = df_encoded["GANANCIA (PÉRDIDA)"] / df_encoded["TOTAL ACTIVOS"]
```

```python
df_encoded.fillna(0, inplace=True)
df_encoded.replace([np.inf, -np.inf], 0, inplace=True)
df_encoded
```

```python
#Visualizar correlaciones antes de crear indice general
import seaborn as sns

correlaciones = df_encoded.corr()
plt.figure(figsize=(10,8)) 
sns.heatmap(correlaciones, cmap="coolwarm") 
plt.show()
```

```python
#Crear Indice General que reemplace al CVI 
peso = [0.3, 0.2, 0.2, 0.3] 
df_encoded['Indice_General'] = df_encoded["RATIO_LIQUIDEZ"] * peso[0] - df_encoded['RATIO_ENDEUDAMIENTO'] * peso[1] + df_encoded['MARGEN_BRUTO'] * peso[2] + df_encoded['ROA'] * peso[3]
df_encoded
```

## Normalizar las columnas numericas

```python
df_encoded = df_encoded.astype(float)
columns = df_encoded.columns.tolist()
for col in columns:
    print(f"Processing column {col}....")
    df_encoded[col] = df_encoded[col].apply(lambda x: (x - df_encoded[col].min()) / (df_encoded[col].max() - df_encoded[col].min()))
                            
df_encoded.head(5)
```
